```python
#see if pairs of features have any connection
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.DataFrame.from_csv(
    'aaf_nums_short_whead.csv',
    index_col=False
)

del df["track_id"]
del df["artist_name"]
del df["song_title"]
df.head()

```

```python
g = sns.FacetGrid(df, hue="mode", palette="Greens_d", size=5)
g.map(plt.scatter, "tempo", "mood_fast", s=50, alpha=.7, linewidth=.5, edgecolor="white")
g.add_legend();
```

```python
g = sns.PairGrid(df)
g.map(plt.scatter)
```
