```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.DataFrame.from_csv(
    'id_key_mode_tempo_mood',
    #index_col=False
)

df_happy = pd.DataFrame.from_csv(
    'id_key_mode_tempo_mood_happy',
    #index_col=False
)

df_sad = pd.DataFrame.from_csv(
    'id_key_mode_tempo_mood_sad',
    #index_col=False
)

df_sad.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
df.hist(column="pos-neg",bins=20, ax=ax)
ax.set_title("Sentiment")
```

```python
alltempo=df["tempo"].values
happytempo=df_happy["tempo"].values
sadtempo=df_sad["tempo"].values
```

```python
plt.hist(alltempo, bins=20, alpha=1, label='All Songs', color="purple", normed=True)
plt.xlabel("Tempo")
plt.legend(loc='upper right')
plt.grid()
plt.show()
```

```python
df_key = pd.DataFrame.from_csv(
    '../sentiment/key_freq',
    index_col=False
)
df_key.head()
```

```python
df_mode= pd.DataFrame.from_csv(
    '../sentiment/mode_freq',
    index_col=False
)

mode=df_mode["mode"].values
freq=df_mode["freq"].values
labels=["minor","major"]
plt.bar(mode, freq, label='All Songs', color="purple", align="center")
plt.legend(loc='upper right')
plt.xlabel("Mode")
plt.xticks(mode, labels)
plt.show()
```

```python
df_key_happy = pd.DataFrame.from_csv(
    '../sentiment/key_freq_happy',
    index_col=False
)

df_key_sad = pd.DataFrame.from_csv(
    '../sentiment/key_freq_sad',
    index_col=False
)
key_happy=df_key_happy["key"].values
freq_happy=df_key_happy["freq"].values/14171.0
key_sad=df_key_sad["key"].values
freq_sad=df_key_sad["freq"].values/12806.0
key_sad = [ x + 0.15 for x in key_sad ]
key_happy = [ x - 0.15 for x in key_happy ]

labels=["C", "C#", "D", "D#", "E", "F", "F#" ,"G" ,"G#" ,"A", "A#" ,"B"]
plt.bar(key_happy, freq_happy, label='POS', color="red", align="center", width=0.25)
plt.bar(key_sad, freq_sad, label='NEG', color="blue", align="center", width=0.25)
plt.legend(loc='upper right')
plt.xlabel("Key")
plt.xticks(key, labels)
plt.show()

```

```python
df_mode_happy = pd.DataFrame.from_csv(
    '../sentiment/mode_freq_happy',
    index_col=False
)

df_mode_sad = pd.DataFrame.from_csv(
    '../sentiment/mode_freq_sad',
    index_col=False
)
mode_happy=df_mode_happy["mode"].values
freq_happy=df_mode_happy["freq"].values/14171.0
#freq_happy=df_mode_happy["freq"].values
mode_sad=df_mode_sad["mode"].values
freq_sad=df_mode_sad["freq"].values/12806.0
#freq_sad=df_mode_sad["freq"].values
mode_sad = [ x + 0.13 for x in mode_sad ]
mode_happy = [ x - 0.13 for x in mode_happy ]

labels=["minor","major"]
plt.bar(mode_happy, freq_happy, label='POS', color="red", align="center", width=0.25)
plt.bar(mode_sad, freq_sad, label='NEG', color="blue", align="center", width=0.25)
plt.legend(loc='upper left')
plt.xlabel("Mode")
plt.xticks(mode, labels)
plt.show()
```
